Данные храняться в типе данных ndarray - n-мерный массив. У него есть размер(shape), тип элементов(dtype) и strides.

Вот базовые операции с array в numpy

```python
import numpy as np
a = np.array([[1,2,3],
                [4,5,6]], dtype=np.float32)

print(a.shape)
print(a.ndim)
print(a.dtype)
print(a.itemsize)
print(a.size)

b = a.reshape(3,2)
print(b)
c = a.ravel()
print(c, c.shape)
d = a.flatten()
print(d)
print(a)
```

```python
X = np.arange(20, dtype=np.float32).reshape(4,5)
#print(X)
print(X[::2, 1:], X)

print()
print(X)
X[X<5]=0
print(X)
np.where(X> 10, X, 0)
```

```python
print(X[1:4][::2])
med = np.median(X, axis=0, keepdims=True)
np.where(X>med, X, 0)

Y = X[[3,0,3,2], :]
Y = Y[:, [4,1]]
print(Y)


print(
X[..., 1::2])
```

Broadcasting - это способность numpy приводить разные массивы к растянутому размера для выполнения поэлементных операций. 

```python
mu = X.mean(axis=0, keepdims=True)
su = X.std(axis=0, keepdims=True)
Z = (X - mu) / (su - 1e-12)
print(Z)

norms = np.linalg.norm(X, axis=1, keepdims=True)
```
